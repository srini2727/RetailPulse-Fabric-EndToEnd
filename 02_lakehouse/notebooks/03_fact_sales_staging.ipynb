```python
from pyspark.sql import functions as F

o = spark.table("silver_orders_clean")
i = spark.table("silver_order_items_clean")

gold_fact_sales_staging = (
    o.join(i, "order_id", "inner")
     .select(
        F.col("order_id").cast("int"),
        F.col("order_ts"),
        F.date_format(F.col("order_date"), "yyyyMMdd").cast("int").alias("date_key"),
        F.col("customer_id").cast("int"),
        F.col("product_id").cast("int"),
        F.col("order_item_id").cast("int"),
        F.col("quantity").cast("int"),
        F.col("unit_price").cast("double"),
        F.col("line_total").cast("double").alias("sales_amount"),
        F.col("channel")
     )
)

gold_fact_sales_staging.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("gold_fact_sales_staging")

print("âœ… gold_fact_sales_staging created")
```
